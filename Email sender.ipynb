# Mail semplice


```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.header import Header
from email.utils import formataddr

email = smtplib.SMTP("smtp.gmail.com", 587)
email.ehlo()
email.starttls()
email.login("mail", "pass")
msg = MIMEMultipart()
msg['From'] = formataddr((str(Header("No reply", 'utf-8')), 'no_reply@nuhuh.com'))
msg['To'] = "dest"
msg['Subject'] = "oggettomail"
msg.attach(MIMEText("testo mail qui", 'plain'))

email.send_message(msg)
email.quit()
```
